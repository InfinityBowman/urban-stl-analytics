```python
import pandas as pd
import json

with open("../../public/data/crime.json", "r") as f:
    crime = json.load(f)
with open("../../public/data/vacancies.json", "r") as f:
    vacancies = json.load(f)

# Extract crime neighborhoods and totals
crime_df = pd.DataFrame(crime["neighborhoods"]).T
crime_df = crime_df.rename(columns={"total": "crime_total"})

# Vacancies with coordinates
vacancies_df = pd.DataFrame(vacancies)
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Convert crime data to GeoDataFrame
crime_gdf = gpd.GeoDataFrame(
    crime_df,
    geometry=[Point(xy) for xy in zip(crime_df.longitude, crime_df.latitude)],
    crs="EPSG:4326",
)

# Convert vacancies data to GeoDataFrame
vacancies_gdf = gpd.GeoDataFrame(
    vacancies_df,
    geometry=[Point(xy) for xy in zip(vacancies_df.longitude, vacancies_df.latitude)],
    crs="EPSG:4326",
)

# Perform spatial join: find nearest vacancy point to each crime neighborhood
merged_gdf = gpd.sjoin_nearest(
    crime_gdf, vacancies_gdf, how="inner", distance_col="dist"
)
```
